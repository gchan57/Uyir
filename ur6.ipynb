```python
import cv2
import numpy as np
from ultralytics import YOLO
import easyocr
from paddleocr import PaddleOCR
from google.colab.patches import cv2_imshow
from google.colab import files

# Load YOLO model
model_path = '/content/drive/MyDrive/best (1).pt'
try:
    model = YOLO(model_path)
except Exception as e:
    print(f"Error loading YOLO model: {e}")

# Initialize OCR readers
reader_easyocr = easyocr.Reader(['en'], gpu=True)  # Enable GPU if available
try:
    reader_paddleocr = PaddleOCR(use_angle_cls=True, lang='en')  # PaddleOCR with GPU if possible
except Exception as e:
    print(f"Error initializing PaddleOCR: {e}")
    reader_paddleocr = None

# Upload images
print("Please upload image files...")
uploaded = files.upload()

# Define class names
class_names = [
    "Airplane", "Car", "bag", "four_wheeler", "helmet", "no_helmet",
    "numberplate", "three_wheeler", "tripple_riding", "two_wheeler",
    "using_mobile", "With Helmet", "Without Helmet", "licence"
]

def preprocess_plate(img, use_grayscale=True):
    if use_grayscale:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.medianBlur(gray, 5)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(blurred)
        thresh = cv2.adaptiveThreshold(enhanced, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                       cv2.THRESH_BINARY_INV, 11, 2)
        kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
        sharpened = cv2.filter2D(thresh, -1, kernel)
        return sharpened
    else:
        kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
        sharpened = cv2.filter2D(img, -1, kernel)
        return sharpened

# Loop through uploaded images
for file_path in uploaded.keys():
    print(f"\nProcessing image: {file_path}")
    img = cv2.imread(file_path)

    try:
        results = model(file_path, conf=0.3)
    except Exception as e:
        print(f"Error during object detection: {e}")
        continue

    for result in results:
        detections = result.boxes.data.cpu().numpy()
        if len(detections) == 0:
            print("No detections found.")
            continue

        classes = result.boxes.cls.cpu().numpy().astype(int)

        for i, box in enumerate(detections):
            x1, y1, x2, y2, conf = map(int, box[:5])
            class_name = class_names[classes[i]]

            if class_name == "numberplate":
                plate_region = img[int(y1):int(y2), int(x1):int(x2)]

                preprocessed_gray = preprocess_plate(plate_region, use_grayscale=True)
                preprocessed_color = preprocess_plate(plate_region, use_grayscale=False)

                print("Preprocessed with Grayscale:")
                cv2_imshow(preprocessed_gray)
                print("Preprocessed without Grayscale:")
                cv2_imshow(preprocessed_color)

                # EasyOCR
                plate_text_easyocr = None
                confidence_easyocr = 0
                try:
                    result_easyocr = reader_easyocr.readtext(
                        preprocessed_gray,
                        allowlist='ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789 ',
                        paragraph=False
                    )
                    print("EasyOCR (Grayscale) Result:", result_easyocr)
                    if result_easyocr:
                        best_result = max(result_easyocr, key=lambda x: x[2])
                        plate_text_easyocr = best_result[1].strip()
                        confidence_easyocr = best_result[2]
                        print(f"EasyOCR Plate: {plate_text_easyocr}, Confidence: {confidence_easyocr}")
                except Exception as e:
                    print(f"EasyOCR Error: {e}")

                # PaddleOCR
                plate_text_paddle = None
                confidence_paddle = 0
                try:
                    paddle_result = reader_paddleocr.ocr(preprocessed_color)
                    print("PaddleOCR (Color) Result:", paddle_result)

                    if paddle_result and isinstance(paddle_result[0], dict):
                        rec_texts = paddle_result[0].get('rec_texts', [])
                        rec_scores = paddle_result[0].get('rec_scores', [])
                        if rec_texts and rec_scores:
                            plate_text_paddle = rec_texts[0]
                            confidence_paddle = rec_scores[0]
                            print(f"PaddleOCR Plate: {plate_text_paddle}, Confidence: {confidence_paddle}")
                except Exception as e:
                    print(f"PaddleOCR Error: {e}")

                # Combine based on confidence
                final_text = None
                if plate_text_easyocr and confidence_easyocr > 0.8:
                    final_text = plate_text_easyocr
                elif plate_text_paddle and confidence_paddle > 0.8:
                    final_text = plate_text_paddle

                # Annotate image
                if final_text:
                    cv2.putText(img, final_text, (int(x1), int(y1)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
                else:
                    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                    cv2.putText(img, "No OCR", (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            elif class_name == "no_helmet":
                cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                cv2.putText(img, class_name, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Show final annotated image
    print("\nFinal Annotated Image:")
    cv2_imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
